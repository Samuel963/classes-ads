```python
from datetime import datetime

def questao_1():
    a = int(input("Digite o primeiro número inteiro: "))
    b = int(input("Digite o segundo número inteiro: "))
    print("A soma é:", a + b)

def questao_2():
    a = float(input("\nDigite o primeiro número: "))
    b = float(input("Digite o segundo número: "))
    print("A subtração é:", a - b)

def questao_3():
    a = float(input("\nDigite o primeiro número real: "))
    b = float(input("Digite o segundo número real: "))
    print("A multiplicação é:", a * b)

def questao_4():
    a = float(input("\nDigite o dividendo: "))
    b = float(input("Digite o divisor: "))
    if b != 0:
        print("O quociente é:", a / b)
    else:
        print("Erro: divisão por zero.")

def questao_5():
    a = int(input("\nDigite o número: "))
    b = int(input("Digite o divisor: "))
    if b != 0:
        print("O resto da divisão é:", a % b)
    else:
        print("Erro: divisão por zero.")

def questao_6():
    a = float(input("\nDigite o primeiro número: "))
    b = float(input("Digite o segundo número: "))
    c = float(input("Digite o terceiro número: "))
    print("A média aritmética é:", (a + b + c) / 3)

def questao_7():
    a = float(input("\nDigite um número: "))
    print("O número ao quadrado é:", a ** 2)

def questao_8():
    a = float(input("\nDigite o primeiro número: "))
    b = float(input("Digite o segundo número: "))
    print("O quadrado da soma é:", (a + b) ** 2)

def questao_9():
    base = float(input("\nDigite a base do triângulo: "))
    altura = float(input("Digite a altura do triângulo: "))
    print("A área do triângulo é:", (base * altura) / 2)

def questao_10():
    preco = float(input("\nDigite o preço do produto: R$ "))
    preco_final = preco - (preco * 0.10)
    print("Preço com 10% de desconto: R$", preco_final)

def questao_11():
    a = float(input("Digite o primeiro número: "))
    b = float(input("Digite o segundo número: "))
    if a > b:
        print("O primeiro número é maior.")
    else:
        print("O primeiro número não é maior.")

def questao_12():
    num = int(input("\nDigite um número: "))
    if num > 0 and num % 2 == 0:
        print("O número é positivo e par.")
    else:
        print("O número não é positivo e par.")

def questao_13():
    idade = int(input("\nDigite sua idade: "))
    if 18 <= idade <= 60:
        print("Idade dentro do intervalo.")
    else:
        print("Idade fora do intervalo.")

def questao_14():
    num = int(input("\nDigite um número: "))
    if num % 3 == 0 or num % 5 == 0:
        print("O número é múltiplo de 3 ou de 5.")
    else:
        print("O número não é múltiplo de 3 nem de 5.")

def questao_15():
    valor = float(input("\nDigite um valor: "))
    if valor != 0:
        print("O valor não é igual a zero.")
    else:
        print("O valor é igual a zero.")

def questao_16():
    num = float(input("\nDigite um número: "))
    if num < 0 or num > 100:
        print("O número está fora do intervalo.")
    else:
        print("O número está dentro do intervalo.")

def questao_17():
    senha1 = input("\nDigite a primeira senha: ")
    senha2 = input("Digite a segunda senha: ")
    if senha1 == senha2:
        print("As senhas são iguais.")
    else:
        print("As senhas são diferentes.")

def questao_18():
    idade = int(input("\nQual sua idade? "))
    tem_cnh = input("Você tem CNH? (s/n): ").lower()
    if idade >= 18 and tem_cnh == "s":
        print("Você pode dirigir.")
    else:
        print("Você não pode dirigir.")

def questao_19():
    texto1 = input("\nDigite o primeiro texto: ")
    texto2 = input("Digite o segundo texto: ")
    if texto1 != texto2:
        print("Os textos são diferentes.")
    else:
        print("Os textos são iguais.")

def questao_20():
    num = float(input("\nDigite um número: "))
    if num < 10 or num > 50:
        print("O número é menor que 10 ou maior que 50.")
    else:
        print("O número está entre 10 e 50.")

def questao_21():
    a = 10
    b = 3.14
    c = True
    d = "Olá mundo!"
    print("Tipos das variáveis:")
    print("a:", type(a))
    print("b:", type(b))
    print("c:", type(c))
    print("d:", type(d))

def questao_22():
    num_str = input("\nDigite um número inteiro (como texto): ")
    num_int = int(num_str)
    print("Convertido para int:", num_int)

def questao_23():
    decimal_str = input("\nDigite um número decimal (como texto): ")
    decimal_float = float(decimal_str)
    print("Convertido para float:", decimal_float)

def questao_24():
    valor_bool = True
    valor_str = str(valor_bool)
    print("\nValor lógico convertido para string:", valor_str)

def questao_25():
    nome = input("\nDigite seu nome: ")
    idade = int(input("Digite sua idade: "))
    altura = float(input("Digite sua altura em metros: "))
    print(f"Nome: {nome}, Idade: {idade} anos, Altura: {altura:.2f} m")

def questao_26():
    num = input("\nDigite um número decimal: ")
    if '.' in num:
        casas_decimais = len(num.split('.')[-1])
        print("Número de casas decimais:", casas_decimais)
    else:
        print("O número não tem casas decimais.")

def questao_27():
    texto = input("\nDigite uma frase: ")
    print("A frase tem", len(texto), "caracteres.")

def questao_28():
    inteiro = int(input("\nDigite um número inteiro: "))
    print("Convertido para float:", float(inteiro))

def questao_29():
    bool1 = input("\nDigite o primeiro valor booleano (True/False): ") == "True"
    bool2 = input("Digite o segundo valor booleano (True/False): ") == "True"
    print("Resultado do AND:", bool1 and bool2)

def questao_30():
    a = input("\nDigite o primeiro número: ")
    b = input("Digite o segundo número: ")
    soma_int = int(float(a)) + int(float(b))
    soma_float = float(a) + float(b)
    if soma_int == soma_float:
        print("As somas como int e float são iguais.")
    else:
        print("As somas como int e float são diferentes.")

def questao_31():
    nome = input("Digite seu nome: ")
    print(f"Olá, {nome}!")

def questao_32():
    a = float(input("\nDigite o primeiro número: "))
    b = float(input("Digite o segundo número: "))
    print("Multiplicação:", a * b)

def questao_33():
    metros = float(input("\nDigite o valor em metros: "))
    print(f"{metros} metros = {metros * 100} centímetros")

def questao_34():
    reais = float(input("\nDigite o valor em reais: R$ "))
    print(f"Equivalente em dólares: US$ {reais / 5.00:.2f}")

def questao_35():
    idade = int(input("\nDigite sua idade: "))
    if idade >= 18:
        print("Maior de idade")
    else:
        print("Menor de idade")

def questao_36():
    dias = int(input("\nDigite a quantidade de dias trabalhados: "))
    print(f"Salário total: R$ {dias * 120:.2f}")

def questao_37():
    preco = float(input("\nDigite o preço do produto: R$ "))
    quantidade = int(input("Digite a quantidade desejada: "))
    print(f"Total a pagar: R$ {preco * quantidade:.2f}")

def questao_38():
    num = int(input("\nDigite um número: "))
    print("Antecessor:", num - 1)
    print("Sucessor:", num + 1)

def questao_39():
    valor = float(input("Digite um valor: "))
    if valor == 100:
        print("O valor é igual a 100.")
    else:
        print("O valor não é igual a 100.")

def questao_40():
    a = float(input("\nDigite o primeiro número: "))
    b = float(input("Digite o segundo número: "))
    print(f"Soma: {a + b}")
    print(f"Subtração: {a - b}")
    print(f"Multiplicação: {a * b}")
    print(f"Divisão: {a / b if b != 0 else 'Divisão por zero'}")

def questao_41():
    nota = float(input("Digite a nota do aluno: "))
    if nota >= 7:
        print("Aprovado")
    else:
        print("Reprovado")

def questao_42():
    idade = int(input("\nDigite sua idade: "))
    if idade >= 16:
        print("Pode votar")
    else:
        print("Não pode votar")

def questao_43():
    temp = float(input("\nDigite a temperatura em °C: "))
    if temp < 18:
        print("Frio")
    elif 18 <= temp <= 30:
        print("Agradável")
    else:
        print("Calor")

def questao_44():
    ano_nasc = int(input("\nDigite o ano de nascimento: "))
    ano_atual = datetime.now().year
    idade = ano_atual - ano_nasc
    if idade >= 18:
        print("Maior de idade")
    else:
        print("Menor de idade")

def questao_45():
    a = float(input("\nDigite o primeiro número: "))
    b = float(input("Digite o segundo número: "))
    if a == b:
        print("Os números são iguais.")
    elif a > b:
        print("O primeiro número é maior.")
    else:
        print("O segundo número é maior.")

def questao_46():
    num = float(input("\nDigite um número: "))
    if num > 0:
        print("Positivo")
    elif num < 0:
        print("Negativo")
    else:
        print("Zero")

def questao_47():
    usuario = input("\nDigite o nome de usuário: ")
    senha = input("Digite a senha: ")
    if usuario == "admin" and senha == "1234":
        print("Acesso permitido")
    else:
        print("Acesso negado")

def questao_48():
    num = int(input("\nDigite um número: "))
    if num % 2 == 0 and num % 3 == 0:
        print("É múltiplo de 2 e de 3")
    else:
        print("Não é múltiplo de 2 e de 3")

def questao_49():
    media = float(input("\nDigite a média final do aluno: "))
    if media >= 9:
        print("Aprovado com louvor")
    elif media >= 7:
        print("Aprovado")
    else:
        print("Reprovado")

def questao_50():
    idade = int(input("\nDigite sua idade: "))
    if 0 <= idade <= 12:
        print("Criança")
    elif 13 <= idade <= 17:
        print("Adolescente")
    elif 18 <= idade <= 64:
        print("Adulto")
    else:
        print("Idoso")

# Mapeamento de questões
questoes = {i: globals()[f"questao_{i}"] for i in range(1, 51)}

def menu():
    while True:
        try:
            print("\n--- MENU DE QUESTÕES ---")
            opcao = int(input("Digite o número da questão (1-50) ou 0 para sair: "))
            if opcao == 0:
                print("Encerrando o programa.")
                break
            elif 1 <= opcao <= 50:
                questoes[opcao]()
            else:
                print("Opção inválida. Escolha entre 1 e 50.")
        except ValueError:
            print("Entrada inválida. Digite um número.")
        except KeyError:
            print("Questão não encontrada.")

menu()
```
